```python
import pandas as pd
import numpy as np
```

```python
all_schools = pd.read_csv('All Schools.csv')
all_agencies = pd.read_excel('All Agencies.xlsx')
us_counties = pd.read_excel('US Counties.xlsx')
county_pop_counts = pd.read_excel('County and Population Counts.xlsx')
incident_count_by_org = pd.read_excel('K12Six Incident Count by Organization.xlsx')
```

```python
all_schools.head()
#all_agencies.head()
#us_counties.head()
#county_pop_counts.head()
#incident_count_by_org.head()
```

```python
breaches = all_schools.merge(incident_count_by_org, on='NCES_ID').reset_index(drop=True)
breaches.sort_values(by='Incident Count', ascending=False).head()
```

```python
breaches[['STATE', 'NCES_ID', 'DISTRICT_NAME', 'Total_Students', 'Incident Count']].drop_duplicates().sort_values(by='Incident Count', ascending=False)

breaches[['STATE', 'Incident Count']].drop_duplicates().fillna(0).groupby('STATE').sum().sort_values(by='Incident Count', ascending=False)
```
